## An example elaboration

### Just look at races where specific competitors won

```python
def single_context_activation_winner(mu, sigma, nrace, winner, max_cycle = 100, threshold = 0.7, init = 0.1, rate = 0.2, sigmoid = True, beta = 35, x_zero = 0.5, cm = True, decay = 0):
    
    actmat_all = np.zeros([nrace, max_cycle + 1, len(mu)])
    
    i = 0
    while i < nrace:
        # Get the activation for one context
        temp_actmat = race.single_race(mu, sigma, max_cycle = max_cycle, threshold = 100, init = init, rate = rate, sigmoid = sigmoid, beta = beta, x_zero = x_zero, cm = cm, decay = decay)     
        if np.argmax(temp_actmat[-1]) == winner:
            actmat_all[i] = temp_actmat
            i = i + 1
        
    return actmat_all
```

Compare acitvation **when the competitor wins the race**

```python
# Get hundred wins by each candidate
a0 = single_context_activation_winner([0.05, 0.045, 0.04], 0.02, 100, 0, cm = True)
a1 = single_context_activation_winner([0.05, 0.045, 0.04], 0.02, 100, 1, cm = True)
a2 = single_context_activation_winner([0.05, 0.045, 0.04], 0.02, 100, 2, cm = True)
# Extract the cases where each competitor wins
winners_array = np.stack((np.mean(a0, axis = 0)[:, 0], np.mean(a1, axis = 0)[:, 1], np.mean(a2, axis = 0)[:, 2]), axis = 1)
plt.plot(winners_array)
```

```python
# Set smaller inhibition rate
irate = 0.05
# Get hundred wins by each candidate
a0 = single_context_activation_winner([0.05, 0.045, 0.04], 0.02, 20, 0, rate = irate, cm = True)
a1 = single_context_activation_winner([0.05, 0.045, 0.04], 0.02, 20, 1, rate = irate, cm = True)
a2 = single_context_activation_winner([0.05, 0.045, 0.04], 0.02, 20, 2, rate = irate, cm = True)
winners_array = np.stack((np.mean(a0, axis = 0)[:, 0], np.mean(a1, axis = 0)[:, 1], np.mean(a2, axis = 0)[:, 2]), axis = 1)
plt.plot(winners_array)
```
